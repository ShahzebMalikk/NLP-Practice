# Nested Code Blocks

```python
word = 'Billionaire'
```

```python
from __future__ import print_function
```

```python
if len(word) > 5:
     print('greater than 5 words');
```

An if statement is known as a control structure because it controls whether the code in the indented block will be run

```python
for word in ['Call', 'me', 'Ishmael', '.']:
    print(word)
```

This is called a loop because Python executes the code in circular fashion

```python
sent1 = ['Call', 'me', 'Ishmael', '.']
```

```python
for xyzzy in sent1:
    if xyzzy.endswith('l'):
        print(xyzzy)
```
